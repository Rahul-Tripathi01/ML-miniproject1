```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
# Train a logistic regression classifier to predict the species of iris
```

```python
iris =datasets.load_iris()
```

```python
print(iris.keys())
```

```python
print(iris['data'])
```

```python
print(iris['target'])
```

```python
print(iris['DESCR'])
```

```python
X=iris['data'][:,:]
y=(iris["target"]).astype(np.int)
```

```python
X,y
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(X,y)
example=clf.predict( [[5.7, 4.4, 1.5, 0.4]])
```

```python
example
```

```python
#using matplotlib for visualization
```

```python
X_new=np.linspace(0,3,1000).reshape(250,4)
y_prob=clf.predict_proba(X_new)
plt.plot(X_new,y_prob[:,1],"g-")
```
